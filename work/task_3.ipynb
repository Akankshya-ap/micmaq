```python
import collections as cl
import os
import sys
import tensorflow as tf
```

```python
def file_to_id(filename,word_to_id):
    with tf.gfile.GFile(filename,'r') as f:
        data=f.read().replace("\n","<eos>").split()
    return [word_to_id[word] for word in data if word in word_to_id]
```

```python
def build_vocab(filename):
    with tf.gfile.GFile(filename,'r') as f:
        data=f.read().replace("\n","<eos>").split()
    counter=cl.Counter(data)
    #print counter
    count_pairs=sorted(counter.items(),key=lambda x:(x[-1],x[0]))
    
    words,_=list(zip(*count_pairs))
    #print words
    word_to_id=dict(zip(words,range(len(words))))
    return word_to_id
word_to_id=build_vocab('/home/apatra/Desktop/work/test.txt')
#print word_to_id
test_d=file_to_id('/home/apatra/Desktop/work/test.txt',word_to_id)
print test_d
```

```python
test_d=tf.convert_to_tensor(test_d,dtype=tf.int32)
tf.Print( test_d,[test_d])
```

```python
def get_data(data_p):
    train_p=os.path.join(data_p,"train.txt")
    valid_p=os.path.join(data_p,"valid.txt")
    test_p=os.path.join(data_p,"test.txt")
    
    word_to_id=build_vocab(train_p)
    train_d=file_to_id(train_p,word_to_id)
    valid_d=file_to_id(valid_p,word_to_id)
    test_d=file_to_id(test_p,word_to_id)
    vocab=len(word_to_id)
    return train_d,valid_d,test_d,vocab
```
